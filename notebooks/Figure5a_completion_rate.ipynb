```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

matplotlib.rc('pdf', fonttype=42)

df = pd.read_parquet('../data/integrated/GANSLIDER_TASKS_TRANSFORMED.parquet')
df
```

```python
deduped = df[['PROLIFIC_PID', 'num_sliders', 'slider_type', 'completed']].drop_duplicates()
deduped
```

```python
counts = deduped.groupby('PROLIFIC_PID').count()
missing_completed_status = deduped[deduped.isna().any(axis=1)].PROLIFIC_PID.unique()
missing_tasks = deduped[(deduped.groupby('PROLIFIC_PID').transform('count') < 14).any(axis=1)].PROLIFIC_PID.unique()

deduped = deduped[(~deduped.PROLIFIC_PID.isin(missing_completed_status))  & (~deduped.PROLIFIC_PID.isin(missing_tasks))].sort_values(['PROLIFIC_PID', 'slider_type', 'num_sliders'])
deduped
```

```python
binom = deduped.copy()
binom['completed'] = (binom['completed'] == "DONE")*1.0
binom
```

```python
completion_rate = deduped.groupby(['num_sliders', 'slider_type']).apply(lambda grp: grp['completed'].value_counts(normalize=True))
completion_rate = completion_rate.reset_index().rename(columns={'level_2': 'completed', 'completed': 'perc'})
completion_rate
```

```python
import pathlib

output = '../r_data/completion_rates.csv'
if not pathlib.Path(output).exists():
	binom.to_csv(output)
```

```python
matplotlib.rc('pdf', fonttype=42)

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

left, width = 0., 1.
bottom, height = 0., 1.
right = left + width
top = bottom + height + .03


test = completion_rate.copy()
test = test.pivot_table(index=['num_sliders'], columns=['slider_type', 'completed'], values='perc')
test.iloc[:, [1, 0, 2, 3]].plot.barh(stacked=True, rot=0, ax=ax, color=['#6CA1AD', '#04677E', '#F3B94E', '#F0D29A'])
ax.set_yticks(list(range(7)))
ax.set_yticklabels([f'{i}' for i in [1, 2, 3, 4, 5, 8, 10]])


custom_lines = [
                Line2D([0], [0], color="#6CA1AD", lw=4),
                Line2D([0], [0], color="#04677E", lw=4),
                Line2D([0], [0], color="#F3B94E", lw=4),
                Line2D([0], [0], color="#F0D29A", lw=4),
                ]

ax.legend(custom_lines, ['Give Up', 'Done', 'Done', 'Give Up'], bbox_to_anchor=(0.95, -.15), fontsize=12, ncol=4)

ax.text(right - .20, top, 'Regular',
        horizontalalignment='right',
        verticalalignment='bottom',
        fontsize=14,
        transform=ax.transAxes)

ax.text(left + .30, top, 'Filmstrip',
        horizontalalignment='right',
        verticalalignment='bottom',
        fontsize=14,
        transform=ax.transAxes)

plt.xlim((0, 2))
plt.xticks(ticks=np.linspace(0, 2, num=9), labels=["1.0", "0.75", "0.5", "0.25", "0", "0.25", "0.5", "0.75", "1.0"])
plt.xlabel('Percentage: Task Success', fontsize=14)
plt.ylabel('Number of Sliders', fontsize=14)

plt.savefig(f'../images/completion_rate.pdf', bbox_inches="tight")
```
