```python
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt

matplotlib.rc('pdf', fonttype=42)
```

```python
df = pd.read_parquet('../data/integrated/GANSLIDER_TASKS_TRANSFORMED.parquet')
df = df[['PROLIFIC_PID', 'taskid', 'num_sliders', 'slider_type', 'completed', 'duration_ms']].drop_duplicates()
df = df.dropna()
df['duration_ms'] = df['duration_ms'].astype(int)
df['duration_sec'] = df['duration_ms'] / 1000
df
```

```python
df.groupby(['num_sliders', 'slider_type']).describe()
```

```python
bar = sns.boxplot(data=df[df.completed == "DONE"], x="num_sliders", y="duration_sec",  hue='slider_type', hue_order=['FILMSTRIP', 'REGULAR'], palette=["C0", "C1"], showfliers=False)

# Define some hatches
hatches = ['///', '']

# Loop over the bars
for i,box in enumerate(bar.artists):
    box.set_edgecolor('#F3B94E'if i%2 else "#04677E" )
    box.set_facecolor('#F3B94E'if i%2 else "white" )
    
    # Set a different hatch for each bar
    box.set_hatch(hatches[ i % 2])

for j, line in enumerate(bar.lines):
    if j % 5 != 4:
        line.set_color("#C6C6C6")

plt.xlabel('Number of Sliders', fontsize=12)
plt.ylabel('Completion Time in Seconds', fontsize=12)

patches = [
    matplotlib.patches.Patch( edgecolor="#04677E", hatch=r'///',label='Filmstrip', fill=False),
    matplotlib.patches.Patch( facecolor='#F3B94E', hatch=r'',label='Regular', fill=True),
]

plt.legend(handles=patches)
plt.savefig('../images/Figure5b_completion_times.pdf', bbox_inches='tight')
```
